```python
import pandas
dataset= pandas.read_excel("genabrev1.xlsx")
import pandas as pd
dataset=pd.read_excel("genabrev1.xlsx")
dataset

from bs4 import BeautifulSoup
import requests

import pandas as pd
i = 0
while i < 11:
    genabrev = pd.read_excel("genabrev1.xlsx", sheet_name = i)
    gen10 = genabrev["Gen Abrev"].to_list()
    i += 1
    n = 0
    while n < len(gen10):
        url = "https://www.uniprot.org/uniprot/?query="+ gen10[n] +"+Bos+Taurus&sort=score"
        n += 1
        html = requests.get(url).text 
        
        soup = BeautifulSoup(html, "lxml")
        
        empty = soup.find("div", {"id" : "noResultsMessage"})
        if empty is not None:
            print("")
            
        else:
        
            entry = soup.find("td", class_="entryID").text

            url_2 = "https://www.uniprot.org/uniprot/"+entry
            html_2 = requests.get(url_2).text

            soup_2 = BeautifulSoup(html_2, "lxml")

            
            protein = soup_2.find("div", {"id" : "content-protein"}).text

            print(protein)
```
